```python
directory = r"A:\Projects\Sanskrit Text Conversion\Sanskrit-Text-Conversion\Data Files\Character_Images"
count = 0
for images in range(len(db)):
    english_literation = db[images][2]
    if english_literation==".":
        continue
    sanskrit_translation = character_map[english_literation]
    if sanskrit_translation=="!":
        continue
    if sanskrit_translation in ["ः", "", " "]:
        print(images)
        count+=1
        continue
    grayImage = cv2.cvtColor(db[images][0], cv2.COLOR_BGR2GRAY)
    string = str(images)
    filename = f"{directory}\{string}.tif"
    cv2.imwrite(filename, grayImage)
    ground_truth_file = f"{directory}\{string}.gt.txt"
    with open(ground_truth_file, "w") as f:
        f.write(sanskrit_translation)
print(count)
```
